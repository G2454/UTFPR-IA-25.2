```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import random
import numpy
from deap import base
from deap import creator
from deap import algorithms
from deap import tools
```

```python
# Gera o toolbox responsável por registrar as configurações do framewrok
toolbox = base.Toolbox()
```

```python
# Cria o tipo de função fitness e indivíduo
creator.create("Maximization", base.Fitness, weights=(1.0,))
creator.create("Genes", list, fitness = creator.Maximization)

# Registra os nomes e os tipos de individuo, fiteness e população
toolbox.register("Atributo", random.randint,0,1)
toolbox.register("Cromossomo", tools.initRepeat, creator.Genes, toolbox.Atributo, n=10  )
toolbox.register("Populacao", tools.initRepeat, list, toolbox.Cromossomo)
```

```python
# Função fitness para 1 objetivo
# Dica: colocar vírgula para manter o padrão com múltiplos onjetivox
def fitnessFunction(chromosome):
   fitness = 0
   for gene in chromosome:
    if gene == 1:
      fitness += 1
   return fitness,
```

```python
# Registra os operadores. Deve-se manter os nomes evaluate, mate, mutate e select
toolbox.register("evaluate", fitnessFunction)
toolbox.register("mate", tools.cxTwoPoint)
toolbox.register("mutate", tools.mutFlipBit, indpb = 0.02)
toolbox.register("select", tools.selTournament, tournsize = 5)
```

```python
# Gera o Algoritmo Genético com a estrutura e operadores cadastrados
pop = toolbox.Populacao(n=20)
prob_cx = 0.8
prob_mut = 0.1
nger = 50
```

```python
finalPop, log = algorithms.eaSimple(pop, toolbox, prob_cx, prob_mut, nger, verbose = False)
```

```python
melhorSolucao = tools.selBest(pop, 1)
for i in melhorSolucao:
  print(f'Indivíduo: {i} - Fitness: {i.fitness}')
```
