```python
#Contrução manual do dicionário

def criaListaAdj():
  listaAdj = {
      'Arad': {'Sibiu': 140, 'Zerind': 75, 'Timisoara': 118},
      'Zerind': {'Arad': 75, 'Oradea': 71},
      'Oradea': {'Zerind': 71, 'Sibiu': 151},
      'Sibiu': {'Arad': 140, 'Oradea': 151, 'Fagaras': 99, 'Rimnicu': 80},
      'Timisoara': {'Arad': 118, 'Lugoj': 111},
      'Lugoj': {'Timisoara': 111, 'Mehadia': 70},
      'Mehadia': {'Lugoj': 70, 'Dobreta': 75},
      'Dobreta': {'Mehadia': 75, 'Craiova': 120},
      'Craiova': {'Dobreta': 120, 'Rimnicu': 146, 'Pitesti': 138},
      'Rimnicu': {'Sibiu': 80, 'Craiova': 146, 'Pitesti': 97},
      'Fagaras': {'Sibiu': 99, 'Bucharest': 211},
      'Pitesti': {'Rimnicu': 97, 'Craiova': 138, 'Bucharest': 101},
      'Bucharest': {'Fagaras': 211, 'Pitesti': 101, 'Giurgiu': 90, 'Urziceni': 85},
      'Giurgiu': {'Bucharest': 90},
      'Urziceni': {'Bucharest': 85, 'Vaslui': 142, 'Hirsova': 98},
      'Hirsova': {'Urziceni': 98, 'Eforie': 86},
      'Eforie': {'Hirsova': 86},
      'Vaslui': {'Iasi': 92, 'Urziceni': 142},
      'Iasi': {'Vaslui': 92, 'Neamt': 87},
      'Neamt': {'Iasi': 87}
  }
  return listaAdj
```

```python
# Função busca em largura
def bfs(listaAdj, cidade, objetivo):

  # Fronteira e lista de visitados
  fronteira = [] #Fila
  visitados = []

  # Adicionar a primeira cidade na fronteira
  fronteira.append(cidade)

  # Enquanto houver cidade na fronteira
  while fronteira:

    # Pegar a primeira cidade da fronteira

    s = fronteira.pop(0)

    # Marcar a cidade s como visitada

    visitados.append(s)

    # Verificar se a cidade s é objetivo
    if s == objetivo:
      break

    # Armazena os vizinhos de s na fronteira

    for vizinho in listaAdj[s].keys():
      if vizinho not in visitados:
        fronteira.append(vizinho)

  # Retorna a lista de cidades visitados
  return visitados
```

```python
cidades = criaListaAdj()
percurso = bfs(cidades, 'Arad', 'Bucharest')
print(percurso)
```
